```python
import re
import numpy as np
import pandas as pd
```

```python
def func_1(x):
    try:
        return float(x)
    except ValueError:
        return np.nan

def find_rating_tulu(x):
    response = re.findall(r'<\|assistant\|> \n(.*)$', x)
    if response:
        rating = re.findall(r'\d', response[0])
        if rating:
            return func_1(rating[0])
        else:
            return np.nan
    else:
        return np.nan
```

```python
p_r1 = (pd.read_csv("politeness_results_p1_tulu2_13b.csv")
            .loc[:, ['a person', 'a male person', 'a female person', 'a white person', 'a black person', 'an Asian person']]
            .rename(columns={'a person':'base',
                             'a male person':'male',
                             'a female person':'female',
                             'a white person':'white',
                             'a black person':'black',
                             'an Asian person':'asian'}))

p_r2 = (pd.read_csv("politeness_predictions_p1_r2_tulu2_13b.csv")
            .loc[:, ['a person', 'a male', 'a female', 'a white person', 'a black person', 'an Asian person']]
            .rename(columns={'a person':'base',
                             'a male':'male',
                             'a female':'female',
                             'a white person':'white',
                             'a black person':'black',
                             'an Asian person':'asian'})
            .applymap(find_rating_tulu))

p_r2.sub(p_r1).describe()
```
