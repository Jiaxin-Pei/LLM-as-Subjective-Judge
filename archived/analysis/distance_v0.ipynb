```python
import re
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import stats
import matplotlib.pyplot as plt
```

# Offensive

```python
offensive = pd.read_table("offensive_results_w.tsv")
```

```python
o_error_base_flant5xxl = -(offensive.loc[:, 'label':'asian_score']).sub(offensive['flant5_xxl_base'], axis=0)
o_error_base_flant5xxl['model'] = 'flan_t5_xxl'
o_error_base_flan_ul2 = -(offensive.loc[:, 'label':'asian_score']).sub(offensive['flan_ul2_base'], axis=0)
o_error_base_flan_ul2['model'] = 'flan_ul2'
o_error_base_llama2 = -(offensive.loc[:, 'label':'asian_score']).sub(offensive['llama2_base'], axis=0)
o_error_base_llama2['model'] = 'llama2'
o_error_base_gpt35 = -(offensive.loc[:, 'label':'asian_score']).sub(offensive['gpt35_base'], axis=0)
o_error_base_gpt35['model'] = 'gpt35'
o_error_base_gpt4 = -(offensive.loc[:, 'label':'asian_score']).sub(offensive['gpt4_base'], axis=0)
o_error_base_gpt4['model'] = 'gpt4'

o_error_base = (pd.concat([o_error_base_flant5xxl, o_error_base_flan_ul2, o_error_base_llama2,
                          o_error_base_gpt35, o_error_base_gpt4],
                          ignore_index=True)
                .rename(columns={'label':'overall_score'}))
o_error_base = pd.melt(o_error_base.reset_index(), id_vars=['index', 'model'],
                       value_name='error', var_name='label')
o_error_base['label'] = o_error_base['label'].apply(lambda x: re.sub('_score', '', x))
```

```python
sns.boxplot(data=o_error_base, x="error", hue="label", col="model", kind="kde")
plt.savefig("plot_error_offensive_base.pdf", format='pdf')
```

```python
o_error_identity_flant5xxl = -(offensive.loc[:, 'label':'asian_score']-(offensive.loc[:,'flant5_xxl_base':'flant5_xxl_asian'].values))
o_error_identity_flant5xxl['model'] = 'flan_t5_xxl'
o_error_identity_flan_ul2 = -(offensive.loc[:, 'label':'asian_score']-(offensive.loc[:, 'flan_ul2_base':'flan_ul2_asian'].values))
o_error_identity_flan_ul2['model'] = 'flan_ul2'
o_error_identity_llama2 = -(offensive.loc[:, 'label':'asian_score']-(offensive.loc[:, 'llama2_base':'llama2_asian'].values))
o_error_identity_llama2['model'] = 'llama2'
o_error_identity_gpt35 = -(offensive.loc[:, 'label':'asian_score']-(offensive.loc[:, 'gpt35_base':'gpt35_asian'].values))
o_error_identity_gpt35['model'] = 'gpt35'
o_error_identity_gpt4 = -(offensive.loc[:, 'label':'asian_score']-(offensive.loc[:, 'gpt4_base':'gpt4_asian'].values))
o_error_identity_gpt4['model'] = 'gpt4'

o_error_identity = (pd.concat([o_error_identity_flant5xxl, o_error_identity_flan_ul2, o_error_identity_llama2,
                          o_error_identity_gpt35, o_error_identity_gpt4],
                          ignore_index=True)
                .rename(columns={'label':'overall_score'}))
o_error_identity = pd.melt(o_error_identity.reset_index(), id_vars=['index', 'model'],
                       value_name='error', var_name='label')
o_error_identity['label'] = o_error_identity['label'].apply(lambda x: re.sub('_score', '', x))
```

```python
sns.displot(data=o_error_identity, x="error", hue="label", col="model", kind="kde")
plt.savefig("plot_error_offensive_identity.pdf", format='pdf')
```
